```python
main_list=[text_1,text_2,text_3]

main_list
```

```python
vectorizer=CountVectorizer()

vec=vectorizer.fit_transform(main_list)

vocab=vectorizer.vocabulary_



vocab.items()
```

```python
vec.toarray()

df=pd.DataFrame(vec.toarray())

column=[]
for i in df.columns:
  for j in vocab.items():
    if i==j[1]:
      column.append(j[0])

df.columns=column

df.sum().reset_index()

```
