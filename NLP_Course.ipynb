```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
```

```python
bow=CountVectorizer(ngram_range=(2,2))
```

```python
corpus=['فیلم برادران لیلا خیلی باحال بود',
        'برادران لیلا چرت و پرت ترین فیلمی بود که دیدم',
        'برادران لیلا درباره اوضاع جامعه آمریکا بود',
        'عجب فیلمی بود برادران لیلا',
        'برادران لیلا رو خیلی دوست داشتم',
        'کارگردان برادران لیلا ممنوع‌الکار شده؟',
        'ترانه علیدوستی هم برای برادران لیلا باید ممنوع‌الکار میشد',
        'فقط اون صحنه چک زدن برادران لیلا',
        'آخر برادران لیلا رو نفهمیدم اصلا',
        'برادران لیلا رو فقط بیست دقیقه اولش رو دیدم']
```

```python
bow.vocabulary_
bow.vocabulary_.items()
k=[i for i in bow.vocabulary_.items()]

k

```

```python
x=bow.fit_transform(corpus)
x=x.toarray()
x
```

```python
df=pd.DataFrame(x)
columns_=[]
for i in df.columns:
  for j in k:
    if i==j[1]:
     columns_.append(j[0])

df.columns=columns_

df
```
